```python
#1. Write a program to find the second largest number in a list without using built-in functions like max(). 
l=[32, 41, 22, 153, 457, 53, 16]
for x in range(len(l)):
    lgst=x
    for j in range(x+1,len(l)):
        if(l[j]>l[lgst]):
            lgst=j
    l[x],l[lgst]=l[lgst],l[x]   
print(l[1])  
```

```python
#2. How would you remove duplicates from a list while maintaining the original order? 
lis=[2,1,2,3,1,2,3,3,4,5]
i=0
for x in lis:
    y=i+1
    while(y<len(lis)):
        if(x==lis[y]):
            lis.pop(y)
        else:
            y+=1
    i+=1
print(lis)
```

```python
#5. Given two lists, write a program to find common elements. 
l1=[1,2,3,4,5,6,7,8,8]
l2=[2,5,2,6,8,9,4,2]
common=[]
for x in l1:
    y=0
    while y<len(l2):
        if(x==l2[y]) and x not in common:
            common+=[x]
            y+=1
        else:
            y+=1
print(f"The common elements in the 2 lists are {common}")
```

```python
#7. Write a Python program to count frequency of each element in a list. 
l=[1,2,3,4,2,4,2,5,1,3,4,5,5]
fre=[]
l1=list(set(l))
for x in range(len(l1)):
    freq=0
    for y in range(len(l)):
        if(l1[x]==l[y]):
            freq+=1
    fre+=[freq]
print(f"The frequency of {l1} in the {l} is {fre}")
```
