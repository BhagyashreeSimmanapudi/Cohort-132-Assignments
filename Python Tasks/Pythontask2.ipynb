# Questions on Loop

```python
#Ques.1. Print numbers from 1 to 10. If a number is even, break the loop using a for loop and else clause. 
for x in range(1,11):
    if(x%2==0):
        break
    else:
        print(x)
```

```python
#Ques.2. Print numbers from 1 to 5, except 3 using a for loop and continue statement.
l=[]
for x in range(1,6):
    if(x==3):
        continue
    else:
        l+=[x]
print(l)
```

```python
#Ques.3. Print numbers in a list until a negative number is encountered using a while loop. 
l=[1,-2,5,4,3,2,1,0,2,3,4,2,2,9,8]
x=0
while (x<len(l)):
    if(l[x]>=0):
        print(l[x])
        x+=1
    else:
        break
```

```python
#Ques.4. Find the common elements in two lists using a for loop. 
list1 = [1,2,3,4,5] 
list2 = [3,4,5,6,7] 
l=[]
for x in list1:
    for y in list2:
        if(x==y):
            l+=[x]
print(l)
 
```

```python
#Ques.5. Print the Fibonacci sequence up to the 10th term using a while loop.
x=f=0
s=1
l=[f,s]
while x<8:
    total=f+s
    f,s=s,total
    x+=1
    l+=[total]
print(f"The fibonacci sequence up to the {len(l)} terms is {l}")
```

```python
#Ques.6. Print numbers divisible by 3 or 5 from 1 to 20 using a for loop. 
l=[]
for x in range(1,21):
    if(x%3==0 or x%5==0):
        l+=[x]
print(l)
```

```python
#Ques.7. Find all prime numbers between 1 and 50 using nested for and if.
l=[]
for x in range(1,51):
    count=0
    for y in range(1,x+1):
        if(x%y==0):
            count+=1
    if(count==2):
        l+=[x]
print(l)
```

```python
#Ques.8. Find the first occurrence of a number in a list using a while loop. 
l=[2,3,1,4,6,3,2,1,6]
x,num=0,6
while x<len(l):
    if(l[x]==num):
        print(f"The first occurrece of {num} in {l} is at position {x+1}")
        break
    x+=1    
```

```python
#Ques.9. Calculate factorial of a number using a while loop. 
while True:
    ipt=input("Enter number")
    if(ipt.isdigit()):
        ipt=int(ipt)
        break
    else:
        print("Number value is only supported")
num,fact=ipt,1
while ipt>=1:
    fact*=ipt
    ipt-=1
print(f"The factorial of {num} is {fact}")
```

```python
#Ques.10. Print below patterns using nested for loops. 
'''(i)  
*
* * 
* * * 
* * * * 
* * * * * 
(ii)
* * * * *  
* * * *  
* * *  
* *  
* 
(iii)  
     *  
    * *  
   * * *  
  * * * *  
 * * * * *  
* * * * * *'''
for i in range(1,6):
    for j in range(1,i+1):
        print("*", end=" ")
    print()
print()
for i in range(5,0,-1):
    for j in range(1,i+1):
        print("*", end=" ")
    print()
print()
count=6
for i in range(1,count+1):
    print(" "*(count-i), end="")
    for j in range(1,i+1):
        print("*", end=" ")
    print()
```
