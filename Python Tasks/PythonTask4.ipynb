```python
import pandas as pd
import numpy as np
import random

# Ques.1. Write a Pandas program to create a DataFrame from a dictionary of lists, 
'''where: 
• One column contains random integers between 10 and 100 (size=10). 
• Another column contains random float values between 0 and 1. 
• Extract the rows where the integer column values are divisible by 5 and the 
float column values are greater than 0.5.'''
ints = []
floats = []
for i in range(10):
    ints.append(random.randint(10,100))
    floats.append(random.random())
df1 = pd.DataFrame({"num":ints,"val":floats})
rows = []
for i in range(len(df1)):
    if df1.loc[i,"num"] % 5 == 0 and df1.loc[i,"val"] > 0.5:
        rows.append(df1.loc[i])
print("Filtered rows are\n", pd.DataFrame(rows))
```

```python
# Ques.2. Given a DataFrame with missing values: 
'''data = {'A': [1, 2, np.nan, 4, 5], 
'B': [5, np.nan, np.nan, 8, 10], 
'C': ['a', 'b', np.nan, 'd', 'e']} 
df = pd.DataFrame(data) 
•  Fill missing values in column A with the mean of the column. 
•  Replace missing values in column B with the median. 
•  Drop rows where column C has missing values.'''
data = {'A': [1, 2, np.nan, 4, 5],
'B': [5, np.nan, np.nan, 8, 10],
'C': ['a', 'b', np.nan, 'd', 'e']}
df = pd.DataFrame(data)
mean=df['A'].mean()
df['A']=df['A'].fillna(mean)
median_b=df['B'].median()
df['B']=df['B'].fillna(median_b)
df=df.dropna(subset=['C'])
print (df)
```

```python
# Ques.3. Given a DataFrame of sales data: 
'''data = {'Region': ['North', 'South', 'North', 'South', 'East'], 
'Product': ['A', 'A', 'B', 'B', 'C'], 
'Sales': [200, 150, 400, 300, 100]} 
df = pd.DataFrame(data) 
•  Pivot the table to show total sales for each product across regions. 
•  Add a new column to the pivot table that calculates the percentage of total sales for each 
product.'''
data = {"Region":["North","South","North","South","East"],
        "Product":["A","A","B","B","C"],
        "Sales":[200,150,400,300,100]}
df = pd.DataFrame(data)
print(df)
pivot_table=df.pivot_table(index='Product',columns='Region',values='Sales',aggfunc='sum')
pivot_table['Total_Sales']=pivot_table.sum(axis=1)
print(pivot_table,'\n')
pivot_table['Percentage']=pivot_table['Total_Sales']/pivot_table['Total_Sales'].sum()*100
print(pivot_table,'\n')
```

```python
# Ques.4.  Write a custom function that: 
'''• Takes a numeric value as input and returns 'High' if the value is above the mean of 
the column, otherwise 'Low'. 
• Apply this function to a column of random integers and add the results as a new 
column. '''
nums = [random.randint(10,100) for i in range(10)]
df4 = pd.DataFrame({"score":nums})
mean_val = sum(df4["score"])/len(df4["score"])
levels = []
for x in df4["score"]:
    if x > mean_val:
        levels.append("High")
    else:
        levels.append("Low")
df4["level"] = levels
print("Scores with High/Low:\n",df4)
```

```python
# Ques.5. Create a Pandas Series of random integers between 1 and 100. 
'''Perform the following tasks: 
1. Create a new series where all values that are greater than the 75th percentile of the 
original series are replaced with the string 'High' and the rest are replaced with 
'Low'. 
2. Compute and display the cumulative sum of the series values.
s = pd.Series([random.randint(1,100) for i in range(15)])'''
random_series = pd.Series(np.random.randint(1, 101, size=10))
print("Original Series:\n", random_series)

percentile_75 = random_series.quantile(0.75)

categorized_series = np.where(random_series > percentile_75, 'High', 'Low')
categorized_series = pd.Series(categorized_series, index=random_series.index)
print("\nCategorized Series (based on 75th percentile):\n", categorized_series)

cumulative_sum = random_series.cumsum()
print("\nCumulative Sum of Original Series:\n", cumulative_sum)
```
