# Part 1: Arithmetic Operators

```python
#Q1: Calculate the total cost when buying 5 books at ₹150 each with 18% GST. 
books = 5 
price = 150 
books,price=5,150
print(f"The total cost when buying {books} books at {price} each with 18% GST is ₹{(books*price)+(books*price)*0.18}")
```

```python
#Q2: Find the average of three test scores: 85, 92, 78 
score1 = 85 
score2 = 92   
score3 = 78
print(f"The average of three test scores: {score1}, {score2}, {score3} is {(score1+score2+score3)/3}")
```

```python
#Q3: Find the remainder when 157 is divided by 12.
dividend,divisor=157,12
print(f"The remainder when {dividend} is divided by {divisor} is {dividend%divisor}.")
```

```python
#Q4: Calculate 2 raised to the power of 8. 
base,power=2,8
print(f"{base} raised to the power of {power} is {base**power}.")
```

```python
#Q5: A worker works 7.5 hours daily. Calculate total minutes. 
hours=7.5
print(f"A worker works {hours} hours daily, ie {hours*60} minutes daily.")
```

```python
#Q13: Check if any vowel exists in "rhythm". 
word = "rhythm" 
vowels = "aeiou"
for x in word:
    if(x in vowels):
        print(f"Vowel exists in {word}")
        break
else:
    print(f"Vowel does not exist in {word}")
```

```python
#Q22: Assign grade based on marks: 
#90+: A, 80-89: B, 70-79: C, 60-69: D, Below 60: F 
marks = 76 
# Write if-elif-else statements 
if(marks>90 and marks<=100):
    print("A")
elif(marks>=80 and marks<90):
    print("B")
elif(marks>=70 and marks<80):
    print("C")
elif(marks>=60 and marks<70):
    print("D")
elif(marks<60):
    print("F")
```

```python
#Q23: Traffic light response: 
#Red: Stop, Yellow: Wait, Green: Go, Other: Invalid 
light = "Red" 
# Write if-elif-else statements 
if(light=="Red"):
    print("Stop")
elif(light=="Yellow"):
    print("Wait")
elif(light=="Green"):
    print("Go")
else:
    print("Invalid")

```

```python
#Q24: Calculate movie ticket price: 
#Age < 12: ₹100, Age 12-60: ₹200, Age > 60: ₹150 
age = 25 
# Write if-elif-else statements 
if(age>0):
    if(age<12):
        print("Price is ₹100")
    elif(age>=12 and age<=60):
        print("Price is ₹200")
    elif(age>60):
        print("Price is ₹150")
else:
    print("Not included")
```

```python
#Q25: Electricity bill calculation: 
#0-100 units: ₹2/unit, 101-200: ₹3/unit, 200+: ₹5/unit 
units = 150 
# Write if-elif-else statements 
if(units>0 and units<=100):
    print(f"The total bill is ₹{units*2}")
elif(units>100 and units<=200):
    print(f"The total bill is ₹{units*3}")
elif(units>200):
    print(f"The total bill is ₹{units*5}")
    
```

```python
#Q26: Check if year is leap year: 
#Divisible by 4 but not by 100, OR divisible by 400 
year = 2024 
# Write if-elif-else statements 
if((year%4==0 and year%100!=0) or year%400==0):
    print("It is a leap year")
```

```python
#Q29: Student result based on marks in 3 subjects: 
math = 75 
science = 68 
english = 82 
total = math + science + english 
percentage = total / 3 
# Use if-elif-else for Pass/Fail (passing = 60%) 
if(percentage>=60 and percentage<=100):
    print("Pass")
elif(percentage>=0 and percentage<60):
    print("Fail")
else:
    print("Invalid Marks")
    
```

```python
#Q30: Discount calculation: 
#Purchase > ₹5000: 20% discount 
#Purchase > ₹2000: 10% discount 
#Purchase > ₹1000: 5% discount 
#Below ₹1000: No discount 
purchase = 3500
# Calculate final amount after discount
if(purchase>5000):
    print(f"Final amount after discount is ₹{(purchase)-((purchase/100)*20)}")
elif(purchase>2000 and purchase<=5000):
    print(f"Final amount after discount is ₹{(purchase)-((purchase/100)*10)}")
elif(purchase>1000 and purchase<=2000):
    print(f"Final amount after discount is ₹{(purchase)-((purchase/100)*5)}")
elif(purchase>0 and purchase<1000):
    print("No discount")
```
